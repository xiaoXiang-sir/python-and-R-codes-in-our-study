```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from scipy.stats import ttest_ind, mannwhitneyu, levene, ks_2samp

# Read Excel files and drop the first column
xlsx_a = r'C:\Users\12292\Desktop\UIA.xlsx'
xlsx_b = r'C:\Users\12292\Desktop\RIA.xlsx'
data_a = pd.read_excel(xlsx_a).iloc[:, 1:]
data_b = pd.read_excel(xlsx_b).iloc[:, 1:]

# Add label column
data_a['label'] = 0
data_b['label'] = 1

# Concatenate data
data = pd.concat([data_a, data_b])

from scipy.stats import ks_2samp, levene, ttest_ind

index = []

for colName in data_a.columns[:]:
    # Check Kolmogorov-Smirnov test and homogeneity of variances
    ks_stat, ks_pval = ks_2samp(data_a[colName], data_b[colName])
    levene_stat, levene_pval = levene(data_a[colName], data_b[colName])
    
    if ks_pval > 0.05 and levene_pval > 0.05: 
        # Homogeneous variances, use independent sample t-test
        t_stat, t_pval = ttest_ind(data_a[colName], data_b[colName])
        if t_pval < 0.05: 
            index.append(colName)  # Add features with significant differences to the list
    else:  
        # Heterogeneous variances, set equal_var=False
        t_stat, t_pval = ttest_ind(data_a[colName], data_b[colName], equal_var=False)
        if t_pval < 0.05: 
            index.append(colName)

# Output number and list of features with significant differences
print("Number of features with significant differences:", len(index))  
print("Features with significant differences:", index)

# Print shapes of the datasets
print("Shape of data_a:", data_a.shape)
print("Shape of data_b:", data_b.shape)
```

```python
# Create two lists to store features with equal and unequal variances
index_equal_var = []
index_unequal_var = []

for colName in data_a.columns[:]:
    # Check Kolmogorov-Smirnov test and homogeneity of variances
    ks_stat, ks_pval = ks_2samp(data_a[colName], data_b[colName])
    levene_stat, levene_pval = levene(data_a[colName], data_b[colName])
    
    if ks_pval > 0.05 and levene_pval > 0.05: 
        # Homogeneous variances, use independent sample t-test
        t_stat, t_pval = ttest_ind(data_a[colName], data_b[colName])
        if t_pval < 0.05: 
            index_equal_var.append(colName)  # Add features with significant differences to the equal variance list
    else:  
        # Heterogeneous variances, set equal_var=False
        t_stat, t_pval = ttest_ind(data_a[colName], data_b[colName], equal_var=False)
        if t_pval < 0.05: 
            index_unequal_var.append(colName)  # Add features with significant differences to the unequal variance list

# Output number and lists of features with significant differences
print("Number of features with equal variance:", len(index_equal_var))  
print("Features with equal variance:", index_equal_var)
print("Number of features with unequal variance:", len(index_unequal_var))  
print("Features with unequal variance:", index_unequal_var)

# Print shapes of the datasets
print("Shape of data_a:", data_a.shape)
print("Shape of data_b:", data_b.shape)
```

```python
# Create four lists to store features, p-values, and information about equal and unequal variances
features_equal_var = []
p_values_equal_var = []
features_unequal_var = []
p_values_unequal_var = []

for colName in data_a.columns[:]:
    # Check Kolmogorov-Smirnov test and homogeneity of variances
    ks_stat, ks_pval = ks_2samp(data_a[colName], data_b[colName])
    levene_stat, levene_pval = levene(data_a[colName], data_b[colName])
    
    if ks_pval > 0.05 and levene_pval > 0.05: 
        # Homogeneous variances, use independent sample t-test
        t_stat, t_pval = ttest_ind(data_a[colName], data_b[colName])
        if t_pval < 0.05: 
            features_equal_var.append(colName)  # Add features with significant differences to the equal variance list
            p_values_equal_var.append(t_pval)   # Store corresponding p-value
    else:  
        # Heterogeneous variances, set equal_var=False
        t_stat, t_pval = ttest_ind(data_a[colName], data_b[colName], equal_var=False)
        if t_pval < 0.05: 
            features_unequal_var.append(colName)  # Add features with significant differences to the unequal variance list
            p_values_unequal_var.append(t_pval)   # Store corresponding p-value

# Output number and lists of features with significant differences and their p-values for equal variance
print("Number of features with equal variance:", len(features_equal_var))  
print("Features with equal variance:", features_equal_var)
print("P-values for equal variance:", p_values_equal_var)

# Output number and lists of features with significant differences and their p-values for unequal variance
print("Number of features with unequal variance:", len(features_unequal_var))  
print("Features with unequal variance:", features_unequal_var)
print("P-values for unequal variance:", p_values_unequal_var)

# Print shapes of the datasets
print("Shape of data_a:", data_a.shape)
print("Shape of data_b:", data_b.shape)
```

```python
# Output features with significant differences and their p-values for equal variance
print("Number of features with equal variance:", len(features_equal_var))  
print("Features with equal variance and their corresponding p-values:")
for i in range(len(features_equal_var)):
    feature_name = features_equal_var[i]
    mean_a = data_a[feature_name].mean()
    std_a = data_a[feature_name].std()
    mean_b = data_b[feature_name].mean()
    std_b = data_b[feature_name].std()

    print(feature_name, ":", p_values_equal_var[i])
    print("Dataset A -", feature_name, ":", "Mean:", mean_a, "± SD:", std_a)
    print("Dataset B -", feature_name, ":", "Mean:", mean_b, "± SD:", std_b)

# Output features with significant differences and their p-values for unequal variance
print("\nNumber of features with unequal variance:", len(features_unequal_var))  
print("Features with unequal variance and their corresponding p-values:")
for i in range(len(features_unequal_var)):
    feature_name = features_unequal_var[i]
    mean_a = data_a[feature_name].mean()
    std_a = data_a[feature_name].std()
    mean_b = data_b[feature_name].mean()
    std_b = data_b[feature_name].std()

    print(feature_name, ":", p_values_unequal_var[i])
    print("Dataset A -", feature_name, ":", "Mean:", mean_a, "± SD:", std_a)
    print("Dataset B -", feature_name, ":", "Mean:", mean_b, "± SD:", std_b)

# Print shapes of the datasets
print("\nShape of dataset A:", data_a.shape)
print("Shape of dataset B:", data_b.shape)
```
