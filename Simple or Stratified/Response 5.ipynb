```python
import pandas as pd
import numpy as np

# Read the Excel file
file_path = r'C:\Users\12292\Desktop\SR3\radiomics all.xlsx'
data = pd.read_excel(file_path)


for seed in range(99, 199, 10):
    # Split data in a 7:3 ratio
    train_data = data.sample(frac=0.7, random_state=seed)
    test_data = data.drop(train_data.index)
    
    # Calculate counts of RIA and UIA in training and testing sets
    train_ria = train_data[train_data.iloc[:, 0] == 1].shape[0]
    train_uia = train_data[train_data.iloc[:, 0] == 0].shape[0]
    test_ria = test_data[test_data.iloc[:, 0] == 1].shape[0]
    test_uia = test_data[test_data.iloc[:, 0] == 0].shape[0]
    
    # Calculate ratios of RIA to UIA
    train_ratio = train_ria / train_uia if train_uia != 0 else np.inf
    test_ratio = test_ria / test_uia if test_uia != 0 else np.inf
    
    # Print results
    print(f"Sampling results for random seed {seed}:")
    print(f"Training set - RIA: {train_ria}, UIA: {train_uia}, Ratio: {train_ratio}")
    print(f"Testing set - RIA: {test_ria}, UIA: {test_uia}, Ratio: {test_ratio}")
    print("----------------------")
    
```
